BREAST CANCER CLASSIFICATION WITH ANN

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import tensorflow as tf
import sklearn.datasets
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import confusion_matrix
```

data collection

```python
breast_cancer_data=sklearn.datasets.load_breast_cancer()
breast_cancer_data
```

```python
df=pd.DataFrame(breast_cancer_data.data,columns=breast_cancer_data.feature_names)
```

```python
df["label"]=breast_cancer_data.target
```

feature & target

```python
X=df.drop(columns="label",axis=1)
y=df["label"]
```

```python
X.head()
```

```python
y.head()
```

train test split

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2,random_state=2)
```

```python
print(X.shape,X_train.shape,X_test.shape)
```

standardize data

```python
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
```

```python
X_train_std=scaler.fit_transform(X_train)
X_test_std=scaler.transform(X_test)
```

### building neural network

```python
import tensorflow as tf
from tensorflow import keras
tf.random.set_seed(3)
```

```python
X_train_std
```

layers of neural network

```python
model=keras.models.Sequential([
    keras.layers.Flatten(input_shape=(30,)),
    keras.layers.Dense(20,activation="relu"),
    keras.layers.Dense(2,activation="sigmoid")
])
```

compiling up the layers of neural network

```python
model.compile(optimizer="adam",
              loss="sparse_categorical_crossentropy",
              metrics=["accuracy"])
```

training the neural network

```python
history=model.fit(X_train_std,y_train,validation_split=0.1,epochs=10)
```

visualizing accuracy and validation accuracy

```python
plt.plot(history.history["accuracy"])
plt.plot(history.history["val_accuracy"])

plt.title("model accuracy")
plt.ylabel("accuracy")
plt.xlabel("epoch")
plt.legend(["trainin data","validation data"],loc="lower right")
plt.show()
```

visualizing loss vs validation lass

```python
plt.plot(history.history["loss"])
plt.plot(history.history["val_loss"])

plt.title("model loss")
plt.ylabel("loss")
plt.xlabel("epoch")
plt.legend(["trainin data","validation data"],loc="lower right")
plt.show()
```

accuracy of the model on test data

```python
loss,accuracy=model.evaluate(X_test_std,y_test)
```

building the predictive system

```python
input_data=(19.69,21.25,130,1203,0.1096,0.1599,0.1974,0.1279,0.2069,0.05999,0.7456,0.7869,4.585,94.03,0.00615,0.04006,0.03832,0.02058,0.0225,0.004571,23.57,25.53,152.5,1709,0.1444,0.4245,0.4504,0.243,0.3613,0.08758)






input_data_as_array=np.asarray(input_data)


#reshape the numpy array as we are predicting for one datapoint
input_data_reshape=input_data_as_array.reshape(1,-1)

#standardize datapoint

input_data_std=scaler.transform(input_data_reshape)

prediction=model.predict(input_data_std)
print(prediction)
prediction_labels=[np.argmax(prediction)]
print(prediction_labels)



if prediction_labels[0]==0:
  print("The tumor is Malignant")
else :
  print("The tumor is Benign")
```
